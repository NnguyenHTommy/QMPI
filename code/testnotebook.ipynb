```python
import qiskit
from qiskit import QuantumCircuit, transpile, QuantumRegister, ClassicalRegister
```

```python
q = QuantumCircuit()
#q.reset(0)
qr1 = QuantumRegister(1, name = "qc1")
qr2 = QuantumRegister(1, name = "qc2")
cr1 = ClassicalRegister(2, name = "cr1")
q.add_register(qr1,qr2)
q.add_register(cr1)
q.x(qr1[0])
q.measure(qr1[0],cr1[0])
q.cx(qr1[0],qr2[0])
q.swap(qr1[0],qr2[0])
q.reset(qr1[0])
q.x(qr1[0])
q.measure(qr1[0],cr1[1])
q.draw('mpl')
```

```python
from qiskit.quantum_info import random_statevector
from qiskit import QuantumCircuit
from qiskit import Aer
aer_sim = Aer.get_backend('aer_simulator')

vec1 = random_statevector(2)
#vec2 = random_statevector(2)
qtest = QuantumCircuit(2)
qtest.initialize(vec1.data,0)
#qtest.initialize(vec2.data,1)
qtest.cx(0,1)
circ = transpile(qtest, aer_sim)
circ.measure_all()
circ.draw('mpl')
```

```python
from qiskit.visualization import plot_histogram
result = aer_sim.run(circ, shots = 10000).result()
counts = result.get_counts(circ)
print(counts)
plot_histogram(counts, title='Bell-State counts')
```
